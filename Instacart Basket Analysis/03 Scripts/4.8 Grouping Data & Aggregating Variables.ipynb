# Answer #6

```python
#Create a spending flag for each user based on the average price
# creating "average_price" column 
df_ords_prods_merge['average_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge.head()
```

If the mean of the prices of products purchased by a user is lower than 10, then flag  as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag as a “High spender.”

```python
#creating flags
#Deriving columns with loc()
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] >= 10, 'spending_flag'] = 'High spender'
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#Check frequency
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# Answer #7

Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
#Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.
#creating "median_days_order" column 
df_ords_prods_merge['median_days_order'] = df_ords_prods_merge.groupby(['user_id'])['days since the last order'].transform(np.median)
```

```python
df_ords_prods_merge.head()
```

If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#creating flags
#Deriving columns with loc()
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_order'] > 20,'order_frequency_flag'] = 'Non-frequent'
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_order'] <= 20,'order_frequency_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_order'] <= 10,'order_frequency_flag'] = 'Frequent customer'
```
